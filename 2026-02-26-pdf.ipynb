## Exercise 1

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wcwidth import width

sns.set(color_codes=True)


salaryRange = np.array(['<300','300-400','400-500','500-600','600-700'])
frequency = np.array([0,200,300,150,250])

#point a
plt.bar(x=salaryRange,height=frequency)
plt.show()

#point b
plt.pie(x=frequency,labels=salaryRange)
plt.show()

#point c
cum_freq = np.cumsum(frequency)
plt.plot(salaryRange,cum_freq)
plt.show()
```
